```julia
using  LinearAlgebra, Statistics, Plots, Distributions, Polynomials
```

```julia
## Initial set up
max_inc = 100.0
min_inc = 1.0
step_size = 0.1
curve = 100;
```

```julia
## Income vector
Income = min_inc:step_size:max_inc;
```

```julia
## Total Utility Vector
TotalUtility = log.(curve, Income);
```

```julia
## Equal sacrifice
sacrefice = .4
@show HI = Income[791]
@show LI = Income[191]
@show HU = TotalUtility[791]
@show LU = TotalUtility[191]
@show HUt = HU - sacrefice
@show LUt = LU - sacrefice
@show HIt = curve^HUt
@show LIt = curve^LUt;
```

```julia
## Calculating the change in income
@show DH = round(HI - HIt, digits = 1)
@show DL = round(LI - LIt, digits =1);
```

## Ability-to-pay

**Equal sacrifice**

If we make each individual sacrefice 0.2 unit of utility, then the tax paid by is:

```julia
@show DH
@show DL;
```

```julia
x1, x2 = [0, HI],     [HI, HI]    # vectors
y1, y2 = [HU, HU],  [HU, 0]    # vectors
x3, x4 = [0, HIt],     [HIt, HIt]    # vectors
y3, y4 = [HUt, HUt],  [HUt, 0]    # vectors
x5, x6 = [0, LI],     [LI, LI]    # vectors
y5, y6 = [LU, LU],  [LU, 0]    # vectors
x7, x8 = [0, LIt],     [LIt, LIt]    # vectors
y7, y8 = [LUt, LUt],  [LUt, 0]    # vectors
plot(Income, TotalUtility, xlabel = "Income", ylabel = "Total Utility", label = "Diminishing Marginal Utility")
plot!([(x1,y1), (x2,y2)], lc=:black, label = "", ls=:dash)
plot!([(x3,y3), (x4,y4)], lc=:black, label = "", ls=:dash)
plot!([(x5,y5), (x6,y6)], lc=:black, label = "", ls=:dash)
plot!([(x7,y7), (x8,y8)], lc=:black, label = "", ls=:dash)
plot!(legend=:outerbottom, legendcolumns=3)
```

**Proportional sacrifice**

If we make each individual lose 20\% of their utility. 

```julia
t = 0.4
@show HI = Income[791]
@show LI = Income[191]
@show HU = TotalUtility[791]
@show LU = TotalUtility[191]
@show HUt = HU * (1 - t)
@show LUt = LU * (1 - t)
@show HIt = curve^HUt
@show LIt = curve^LUt
@show DH = HI - HIt
@show DL = LI - LIt;
```

```julia
@show DH
@show DL;
```

```julia
x1, x2 = [0, HI],     [HI, HI]    # vectors
y1, y2 = [HU, HU],  [HU, 0]    # vectors
x3, x4 = [0, HIt],     [HIt, HIt]    # vectors
y3, y4 = [HUt, HUt],  [HUt, 0]    # vectors
x5, x6 = [0, LI],     [LI, LI]    # vectors
y5, y6 = [LU, LU],  [LU, 0]    # vectors
x7, x8 = [0, LIt],     [LIt, LIt]    # vectors
y7, y8 = [LUt, LUt],  [LUt, 0]    # vectors
plot(Income, TotalUtility, xlabel = "Income", ylabel = "Total Utility", label = "Diminishing Marginal Utility")
plot!([(x1,y1), (x2,y2)], lc=:black, label = "", ls=:dash)
plot!([(x3,y3), (x4,y4)], lc=:black, label = "", ls=:dash)
plot!([(x5,y5), (x6,y6)], lc=:black, label = "", ls=:dash)
plot!([(x7,y7), (x8,y8)], lc=:black, label = "", ls=:dash)
plot!(legend=:outerbottom, legendcolumns=3)
```

**Comparisons with the absolute utility**

If we go back to the equal absolute sacrifice, lets assume the same drop in utility for the high-income person (red). The used to lose DL = 9.014394566938822.


```julia
@show HU - HUt
@show LUt = LU - (HU - HUt)  # new absolute change for the low income person
@show LIt = curve^LUt
@show DL = LI - LIt;
```

Now they lose 11.674467925981268... In other words, the equal proportional sacrifice is more progressive.
