```julia
using  LinearAlgebra, Statistics, Plots, Distributions, Polynomials
```

```julia
## Initial set up
max_inc = 100.0
min_inc = 1.0
step_size = 0.1
curve = 100;
```

```julia
## Income vector
Income = min_inc:step_size:max_inc;
```

```julia
## Total Utility Vector
TotalUtility = log.(curve, Income);
```

```julia
plot(Income, TotalUtility, xlabel = "Income", ylabel = "Total Utility", label = "Diminishing Marginal Utility")
```

```julia
## Equal sacrefice
sacrefice = .2
@show HI = Income[800]
@show LI = Income[200]
@show HU = TotalUtility[800]
@show LU = TotalUtility[200]
@show HUt = HU - sacrefice
@show LUt = LU - sacrefice
@show HIt = curve^HUt
@show LIt = curve^LUt;
```

```julia
## Calculating the change in income
DH = round(HI - HIt, digits = 1)
DL = round(LI - LIt, digits =1);
```

## Ability-to-pay

**Equal sacrefice**

If we make each individual sacrefice 0.2 unit of utility, then the tax paid by is:

```julia
@show DH
@show DL;
```

**Proportional sacrefice**

If we make each individual lose 20\% of their utility. 

```julia
t = 0.2
@show DH = HI - curve^(HU * (1 - t))
@show DL = LI - curve^(LU * (1 - t))
@show HI
@show HU * (1 - t);
```

```julia
@show DH
@show DL;
```
